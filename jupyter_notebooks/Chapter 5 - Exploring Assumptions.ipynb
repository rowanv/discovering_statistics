### Exploring Assumptions

The key assumptions of parametric tests:
1. Normally distributed data
2. Homogeneity of variance
3. Interval Data
4. Independence



```python
    - Testing for normality:
        - A Q-Q plot (quantile-quantile plot) -- tells us the proportion of cases we find below a certain value. We plot the cumulative values we have in the data against the cumulative probability of a particular distribution (normal for our instance)
            If they are normally distributed, we would essentialy have a diagonal line
        - Shapiro-Wilk test: lets us test whether a distribution si normal 
    - Testing for homogeneity of variance 
```

```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np
import scipy
```

# 5.5 The assumption of normality 

```python
DATA_URLS = {
    'Festival': 'https://studysites.uk.sagepub.com/dsur/study/DSUR%20Data%20Files/Chapter%205/DownloadFestival.dat',
}

```

```python
df_festival = pd.read_csv(DATA_URLS['Festival'], sep='\t')
```

```python
df_festival = df_festival[df_festival['day1'] <= 10]
```

```python
df_festival.head()
```

```python
ax = sns.histplot(df_festival['day1'], stat='density')

# From: https://stackoverflow.com/questions/52908925/how-to-add-a-standard-normal-pdf-over-a-seaborn-histogram

# Add in normal distribution
# function for mapping the pdf

mean = df_festival['day1'].mean()
sd = df_festival['day1'].std()
x = np.random.normal(mean, sd, size=100000)

ax = sns.kdeplot(x)
```

#### Let's look at the QQ plot

A Q-Q plot (quantile-quantile plot) -- tells us the proportion of cases we find below a certain value. We plot the cumulative values we have in the data against the cumulative probability of a particular distribution (normal for our instance)

If they are normally distributed, we would essentialy have a diagonal line

```python
from scipy.stats import probplot
```

```python
probplot(df_festival['day1'], dist='norm', plot=plt)
plt.show()
sns.despine()
```
